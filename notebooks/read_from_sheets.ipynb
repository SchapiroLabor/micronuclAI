```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

```

```python
sheet_id = "1IV3YtMCBOBag32ZFbUQx3Z8yWT8l391x2HV-wfda8o4"
sheet_name = "H358"
url = f"https://docs.google.com/spreadsheets/d/{sheet_id}/gviz/tq?tqx=out:csv&sheet={sheet_name}"
df = pd.read_csv(url)
```

```python
df
```

```python
df = df.dropna()
df
```

```python
# Create micronuclei_cell_ratio for both manual and stardist
df["micronuclei_cell_ratio_manual"] = df["micronuclei_manual"]/df["cells_manual"]
df["micronuclei_cell_ratio_stardist"] = df["micronuclei_stardist_prediction"]/df["cells_stardist_prediction"]
df
```

```python
# Melt the dataframe
df = df.melt(id_vars=["file", "modification"], value_vars=["micronuclei_cell_ratio_manual", "micronuclei_cell_ratio_stardist"], var_name="micronuclei_cell_ratio_type", value_name="micronuclei_cell_ratio")
df
```

```python
sns.catplot(data=df,y="file", x="micronuclei_cell_ratio", hue="micronuclei_cell_ratio_type", col="modification", kind="bar")
plt.show()
```

```python
sns.catplot(data=df, y="modification", x="micronuclei_cell_ratio", hue="micronuclei_cell_ratio_type", kind="box")
plt.show()
```
